```python
import matplotlib.pyplot as plt

## Line Graph Solutions

# Provided data
years = [2010, 2011, 2012, 2013, 2014]
population = [100000, 120000, 140000, 160000, 180000]

plt.plot(years, population)
plt.show()


# Use the same data as in Task 1

plt.plot(years, population)
plt.xlabel('Year')
plt.ylabel('Population')
plt.title('Population Over Years with Labels')
plt.show()



# Provided data for multiple cities
city1_population = [100000, 120000, 140000, 160000, 180000]
city2_population = [90000, 110000, 130000, 150000, 170000]

plt.plot(years, city1_population, label='City 1')
plt.plot(years, city2_population, label='City 2')
plt.xlabel('Year')
plt.ylabel('Population')
plt.title('Population Comparison')
plt.legend()
plt.show()
```

```python
## Bar graph solutions

# Provided sales data
products = ['Product A', 'Product B', 'Product C', 'Product D']
sales = [1200, 1500, 900, 2000]

plt.bar(products, sales)
plt.xlabel('Products')
plt.ylabel('Sales')
plt.title('Product Sales')
plt.show()


# Use the same data as in Task 4

colors = ['red', 'green', 'blue', 'purple']

plt.bar(products, sales, color=colors)
plt.xlabel('Products')
plt.ylabel('Sales')
plt.title('Product Sales with Custom Colors')
plt.show()


# Use the same data as in Task 4

plt.barh(products, sales)
plt.xlabel('Sales')
plt.ylabel('Products')
plt.title('Product Sales (Horizontal)')
plt.show()

```

```python
## histogram solutions

# Provided exam score data
exam_scores = [78, 85, 92, 88, 75, 68, 94, 82, 90, 72, 98, 85, 88, 76, 89]

plt.hist(exam_scores, bins=10, edgecolor='black')
plt.xlabel('Exam Score')
plt.ylabel('Frequency')
plt.title('Exam Score Distribution')
plt.show()


# Use the same data as in Task 7

plt.hist(exam_scores, bins=5, edgecolor='black')
plt.xlabel('Exam Score')
plt.ylabel('Frequency')
plt.title('Exam Score Distribution (Different Bins)')
plt.show()



# Sample data for subplots
data1 = [1, 4, 9, 16]
data2 = [1, 8, 27, 64]
# Create subplots
fig, axes = plt.subplots(1, 2)

# Plotting data in subplots
axes[0].plot(data1)
axes[1].plot(data2)

# Set titles for subplots
axes[0].set_title('Plot 1')
axes[1].set_title('Plot 2')

plt.show()
```

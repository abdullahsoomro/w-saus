**Data visualization:** It is the graphical representation of information and data. It involves the use of visual elements such as charts, graphs, maps, and other graphical formats to help people understand and interpret data more effectively. The primary goal of data visualization is to convey complex information in a clear and concise manner, allowing viewers to quickly grasp patterns, trends, and insights that might be difficult to discern from raw data alone.

### Benefits of Data Visualization

1. **Easy Understanding**: Visualizing data makes it easier to understand complex information by representing it in a graphical format.

2. **Insight Discovery**: Visualizations can reveal hidden patterns, trends, and correlations that may not be apparent in raw data.

3. **Effective Communication**: Visuals are more engaging and can convey information more effectively than raw numbers or text alone, making it easier to communicate findings to stakeholders.

4. **Decision-Making**: Clear and intuitive visualizations enable faster and more informed decision-making based on a better understanding of the data.

5. **Anomaly Idendification**: Visualizations can help in identifying outliers or anomalies in the data, which may require further investigation.

6. **Comparative Analysis**: Visual representations allow for easy comparison between different data sets, making it easier to identify trends and outliers across various categories.

7. **Improved Reporting**: Visualizations are often used in reports and presentations to convey information in a more engaging and memorable way, enhancing the impact of the message.


## Matplotlib
Matplotlib is a widely-used Python library for creating static, animated, and interactive visualizations in a variety of formats. It provides a high-level interface for drawing attractive and informative statistical graphics. Matplotlib is particularly well-suited for creating plots such as line plots, scatter plots, bar plots, error bars, histograms, and more.



## Pyplot
pyplot is a module within the Matplotlib library, which provides a MATLAB-like interface for creating a wide variety of static plots. It is one of the most commonly used modules in Matplotlib for generating visualizations.

### Types of Graphs

- **Line Graphs**:
  - Used to display data points over a continuous interval or time span.
  - Ideal for showing trends or changes in data over time.
  - Points are connected by lines to visualize the progression.

- **Bar Charts**:
  - Represent data with rectangular bars, where the length or height of each bar corresponds to the value of the data it represents.
  - Typically used for comparing discrete categories.

- **Histograms**:
  - Display the distribution of a dataset.
  - Similar to a bar chart, but the data is grouped into intervals or bins.

- **Scatter Plots**:
  - Display individual data points on a two-dimensional graph.
  - Useful for showing the relationship between two continuous variables.

- **Pie Charts**:
  - Represent parts of a whole.
  - Circular graph where each "slice" represents a percentage or proportion of the total.

- **Box-and-Whisker Plots (Boxplots)**:
  - Display the distribution of a dataset, showing the median, quartiles, and potential outliers.



## Exploratory Data Analysis (EDA)

Exploratory Data Analysis (EDA) is an approach to analyzing datasets to summarize their main characteristics, often utilizing visual methods. It is a crucial step in the data analysis process as it helps in understanding the underlying structure, patterns, and relationships within a dataset before applying more complex modeling techniques.

The main objectives of EDA include:

1. **Understanding the Data**: EDA helps in getting a first-hand understanding of the data, its variables, and their relationships.

2. **Detecting Patterns and Anomalies**: EDA helps in identifying patterns, trends, outliers, and anomalies within the data.

3. **Formulating Hypotheses**: By exploring the data visually, analysts can develop initial hypotheses about relationships or trends that may exist in the data.

4. **Feature Engineering**: EDA often leads to the creation of new features or transformations of existing ones to better represent underlying patterns.

5. **Data Preparation**: It can guide decisions on data cleaning, handling missing values, and transforming variables.

6. **Model Selection**: EDA can provide insights into which modeling techniques may be most appropriate for the data.

7. **Communication**: Visualizations and summaries produced during EDA can be used to communicate findings to stakeholders.

EDA techniques often involve calculating summary statistics, generating various types of plots and visualizations, handling missing data, and exploring relationships between variables. It is an iterative process and is conducted throughout the data analysis workflow.

```python
import matplotlib.pyplot as plt

days = ["Mon", "Tue", "Wed", "Thu", "Fri", "Sat", "Sun"]
steps_walked = [8934, 14902, 3409, 25672, 12300, 2023, 6890]

plt.plot(steps_walked)
plt.show()
```

### Adding titles and labels


```python
plt.plot(days, steps_walked, "+-g")
plt.title("Step count for the week")
plt.xlabel("Days of the week")
plt.ylabel("Steps walked")
plt.show()
```

```python
steps_last_week = [9788, 8710, 5308, 17630, 21309, 4002, 5223]

plt.plot(days, steps_walked, "o-g")
plt.plot(days, steps_last_week, "v--m")

plt.title("Step count | This week and last week")
plt.xlabel("Days of the week")
plt.ylabel("Steps walked")
# plt.grid(True)
plt.show()
```

### Creating Subplots


```python

fig, axs = plt.subplots(1, 2)

axs[0].plot(days, steps_walked, "o-g")
axs[0].plot(days, steps_last_week, "v--m")
axs[0].set_title("Step count | This week and last week")
axs[0].set_xlabel("Days of the week")
axs[0].set_ylabel("Steps walked")
axs[0].grid(True)

plt.show()
```

## What’s a Matplotlib Figure Made of?

When working with data visualization in Python, you’ll want to have control over all aspects of your figure.


There are three main parts to a Matplotlib figure:

1. **Figure**:
   - This is the whole region of space that’s created when you create any figure. The Figure object is the overall object that contains everything else.
   - Essentially, it's the canvas on which you'll be creating your plots.

2. **Axes**:
   - An Axes object is the object that contains the x-axis and y-axis for a 2D plot. Each Axes object corresponds to a plot or a graph. You can have more than one Axes object in a Figure, as you’ll see later on in this Chapter.

3. **Axis**:
   - An Axis object contains one of the axes, the x-axis or the y-axis for a 2D plot.
   - For instance, the x-axis represents the horizontal scale, while the y-axis represents the vertical scale.

These components work together to form the structure of a Matplotlib figure, providing a foundation for creating various types of visualizations.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Create a hypothetical dataset
data = {
    'Square_Footage': [1500, 2000, 1800, 2100, 1600, 1900, 2200, 2300, 2500, 2000],
    'Bedrooms': [3, 4, 3, 2, 2, 3, 4, 4, 5, 4],
    'Bathrooms': [2, 3, 2, 3, 1, 2, 3, 3, 3, 2],
    'Sale_Price': [250000, 300000, 280000, 320000, 200000, 270000, 350000, 360000, 400000, 310000]
}

df = pd.DataFrame(data)

# Summary Statistics
summary_stats = df.describe()
print("Summary Statistics:")
print(summary_stats)

# Data Visualization
plt.figure(figsize=(12, 4))

plt.subplot(1, 3, 1)
# plt.hist(df['Square_Footage'])
sns.histplot(df['Square_Footage'], kde=True)
plt.title('Square Footage Distribution')

plt.subplot(1, 3, 2)
sns.boxplot(x='Bedrooms', y='Sale_Price', data=df)
# plt.boxplot((df['Bedrooms'],df['Sale_Price']))
plt.title('Sale Price vs. Bedrooms')

plt.subplot(1, 3, 3)
plt.scatter(df['Square_Footage'],df['Sale_Price'])
# sns.scatterplot(x='Square_Footage', y='Sale_Price', data=df)
plt.title('Square Footage vs. Sale Price')

plt.tight_layout()
plt.show()

# Correlation Analysis
correlation_matrix = df.corr()
print("\nCorrelation Matrix:")
print(correlation_matrix)

# Handling Missing Values
missing_values = df.isnull().sum()
print("\nMissing Values:")
print(missing_values)
```
